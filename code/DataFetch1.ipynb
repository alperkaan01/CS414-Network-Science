```python
from datetime import datetime
from dateutil.relativedelta import relativedelta    

def conv_date(date_str):
  '''This function takes an date string in %Y-%m-%d format 
  and converts that into a date object
  - date_str @string '''
  if(len(date_str) == 4):
    date_str += "-1-1"
  if(len(date_str) == 7):
    date_str += "-1"
  date_object = datetime.strptime(date_str, '%Y-%m-%d').date()
  return date_object

def x_years_later(date,change):
  date = date - relativedelta(years = change)
  return date
```

```python
def is_letter(char):
    ascii_val = ord(char)
    is_uppercase = ascii_val >= 65 and ascii_val <= 90
    is_lowercase = ascii_val >= 97 and ascii_val <= 122
    return is_uppercase or is_lowercase
```

```python
def fix_delimiter_error2(json_string):
    # Pattern to identify extra double quotes within values
    pattern = r'(")(.*?[^\\])(")'

    def replace_quotes(match):
        value = match.group(2)
        value = value.replace('"', "'")
        return f'"{value}"'

    # Find and replace extra double quotes within values with single quotes
    fixed_json_string = re.sub(pattern, replace_quotes, json_string)

    art_ind = fixed_json_string.index(", \"artists\"")
    fixed_json_string = fixed_json_string[:art_ind] + "}"
    for i in range(1,len(fixed_json_string)-2):
      # if (not (fixed_json_string[i-1] == " " or fixed_json_string[i-1] == ":" or fixed_json_string[i-1] == "{" or fixed_json_string[i-1] == "}" or fixed_json_string[i-1] == "," ) and
      #  not (fixed_json_string[i+1] == " " or fixed_json_string[i+1] == ":" or fixed_json_string[i+1] == "{" or fixed_json_string[i+1] == "}") and (fixed_json_string[i+1] == "}") and 
      #  fixed_json_string[i] == "\""):
      if ((is_letter(fixed_json_string[i-1]) and is_letter(fixed_json_string[i+1])) or ((fixed_json_string[i+1] == " " ) and is_letter(fixed_json_string[i+2])) or (fixed_json_string[i+1] == "\""  and fixed_json_string[i+2] == "," ) ) and fixed_json_string[i] == "\"" :

        fixed_json_string = fixed_json_string[:i] + "'" + fixed_json_string[i+1:]   
    try:
        # Parse the modified JSON string
        json_object = json.loads(fixed_json_string)
        return json_object
    except json.JSONDecodeError as e:
        # Handle the JSON decoding error
        print("JSON decoding error:", e)
        return None
```

```python
def fix_delimiter_error(json_string):
    # Pattern to identify and replace single quotes around keys and values
    pattern = r"'([^']+)'"

    # Find and replace single quotes around keys and values with double quotes
    fixed_json_string = re.sub(pattern, r'"\1"', json_string)
   
    try:
        # Parse the modified JSON string
         return fix_delimiter_error2(fixed_json_string)
    except json.JSONDecodeError as e:
        # Handle the JSON decoding error
        print("JSON decoding error:", e)
        return None
```

```python
def check_pop_change(artist,song):
  prev_albums = [album for album in artist["albums"] if conv_date(album["release_date"]) < x_years_later(conv_date(song["release_date"]),-1) ]
  next_albums = [album for album in artist["albums"] if conv_date(album["release_date"]) > x_years_later(conv_date(song["release_date"]),1) ]

  prev_stats = {"album_pop" : 0, "album_count":0, "albums_pop_list":[]}
  next_stats = {"album_pop" : 0, "album_count":0, "albums_pop_list":[]}


  for prev_album in prev_albums:
    prev_stats["albums_pop_list"].append(prev_album["popularity"])
    prev_stats["album_pop"] += prev_album["popularity"]
    prev_stats["album_count"] += 1

  for next_album in next_albums:
    next_stats["albums_pop_list"].append(next_album["popularity"])
    next_stats["album_pop"] += next_album["popularity"]
    next_stats["album_count"] += 1

  isSuccessful = False
  if prev_stats["album_count"] != 0 and next_stats["album_count"] != 0:
    prev_stats["avg"] = prev_stats["album_pop"]/prev_stats["album_count"]
    next_stats["avg"] = next_stats["album_pop"]/next_stats["album_count"]
    if prev_stats["avg"] * 1.1 < next_stats["avg"]:
      isSuccessful = True
      if "successfull_collab" in artist:
        artist["successfull_collab"] += 1
      else:
        artist["successfull_collab"] = 1
  else:
    if "successfull_collab" in artist:
        artist["successfull_collab"] += 0
    else:
        artist["successfull_collab"] = 0

  return (artist , isSuccessful )
```

```python
def get_all_tracks(artists_sample):
  ''' Gets all kind of albums, singles and songs of given artist list'''
  artist_sample["Albums"] = None
  artist_sample["Singles"] = None
  artist_sample["Songs"] = None
  artist_sample.head(10)



  for index, row in artist_sample.iterrows():
    artist_id = row["id"]
    artist_url = "https://api.spotify.com/v1/artists/{}/albums?include_groups=album&limit=50".format(artist_id)  # Replace {artist_id} with the artist's ID
    # Make a request to get the artist's information
    headers = {
        "Authorization": f"Bearer {access_token}"
    }
    response = requests.get(artist_url, headers=headers)
    response_data = response.json()
    album_ids = []
    for response in response_data['items']:
      album_ids.append(response['id'])
    artist_sample.at[index, "Albums"] = album_ids

    artist_url = "https://api.spotify.com/v1/artists/{}/albums?include_groups=single&limit=50".format(artist_id)  # Replace {artist_id} with the artist's ID

    # Make a request to get the artist's information
    headers = {
        "Authorization": f"Bearer {access_token}"
    }
    response = requests.get(artist_url, headers=headers)
    response_data = response.json()

    single_ids = []

    for response in response_data['items']:
      single_ids.append(response['id'])

    artist_sample.at[index, "Singles"] = single_ids

```

```python
def get_songs(artist_sample):
  ''' Gets all songs of given artist list'''
  for index, row in artist_sample.iterrows():
    album_list = row["Albums"]
    single_list = row["Singles"]
    song_ids = []
    for album in album_list:
      album_id = album

      artist_url = "https://api.spotify.com/v1/albums/{}/tracks".format(album_id)  # Replace {artist_id} with the artist's ID

      # Make a request to get the artist's information
      headers = {
          "Authorization": f"Bearer {access_token}"
      }
      response = requests.get(artist_url, headers=headers)
      response_data = response.json()
      for response in response_data['items']:
        song_ids.append(response['id'])
    
    for single in single_list:
      single_id = single

      artist_url = "https://api.spotify.com/v1/albums/{}/tracks".format(single_id)  # Replace {artist_id} with the artist's ID

      # Make a request to get the artist's information
      headers = {
          "Authorization": f"Bearer {access_token}"
      }
      try:
        response = requests.get(artist_url, headers=headers)
        response_data = response.json()
        for response in response_data['items']:
          song_ids.append(response['id'])
      except Exception as e:
        print("Error: ",e)  

    artist_sample.at[index, "Songs"] = song_ids

```
